# Prova Parcial



**Dados:**

São dois arquivos csv:

1. dataset_alunos.csv:
    1. household_key: identificador da família
    2. WEEK_NO_BINNED: identificador do intervalo de 12 semanas (aproximadamente 1 trimestre)
    3. SALES_VALUE: total gasto pela família
    4. COUPON_DISCOUNT: total de desconto aplicado pela utilização de cupons promocionais
    5. COUPON_DISC_BOOL: variavel binária correspondente a coluna COUPON_DISCOUNT
2. hh_demog_alunos.csv: contém diversas caracterizações relacionadas a cada família

**Objetivo:**

O objetivo do exercício será montar um modelo para estimar a variável SALES_VALUE no intervalo (84, 96], ou seja da 85a até a 96a semana.

**Métrica:**

Será utilizada a métrica RMSE

**Arquivo a ser submetido (IMPORTANTE):**

Voces deverão submeter o jupyter notebook junto com o arquivo csv `dataset_alunos.csv` trocando o `"alunos"` pelo seu nome conforme se encontra registrado no aluno online (caixa baixa, retirar acentos e substituir espaços por underscore). Seu arquivo de submissão precisa estar no mesmo formato do arquivo de dados, mas com o valor de predição na <br>
Por exemplo:
> nome: João Silva<br>
> arquivo de submissão: `dataset_joao_silva.csv`

**Outros pontos importantes:**

* Exercicio individual a ser entregue até o dia 19/outubro/2021 as 23:59. 
* Alguns alunos poderão ser convocados para uma arguição oral no dia 22/outubro/2020, durante o horário da aula.
* A nota será proporcional ao desempenho do modelo.
* Utilizar somente os algoritmos abordados no curso. 

**Dicas:**
- funções pd.pivot() e/ou pd.pivot_table()
- dtype 'category' do pandas (pesquisar em https://pandas.pydata.org/pandas-docs/stable/user_guide/categorical.html sobre variaveis com dtype category)

## Target Encoding (Sem contexto Temporal - household_key)

```python
colunas = ['SALES_VALUE',
           'COUPON_DISC']
```

```python
#mediana
dfmedian = df.groupby(['household_key'])[colunas].quantile(0.5).reset_index()

columns = dfmedian.columns[1:]
columns = "median_" + columns
columns = ['household_key'] + list(columns)
dfmedian.columns = columns

df = pd.merge(df, dfmedian, how="left", on=['household_key'])


#média
dfmean = df.groupby(['household_key'])[colunas].mean().reset_index()

columns = dfmean.columns[1:]
columns = "mean_" + columns
columns = ['household_key'] + list(columns)
dfmean.columns = columns

df = pd.merge(df, dfmean, how="left", on=['household_key'])


#percentil 25
df25 = df.groupby(['household_key'])[colunas].quantile(0.25).reset_index()

columns = df25.columns[1:]
columns = "perc25_" + columns
columns = ['household_key'] + list(columns)
df25.columns = columns

df = pd.merge(df, df25, how="left", on=['household_key'])


#percentil 75
df75 = df.groupby(['household_key'])[colunas].quantile(0.75).reset_index()

columns = df75.columns[1:]
columns = "perc75_" + columns
columns = ['household_key'] + list(columns)
df75.columns = columns

df = pd.merge(df, df75, how="left", on=['household_key'])


#std
dfstd = df.groupby(['household_key'])[colunas].std().reset_index()

columns = dfstd.columns[1:]
columns = "std_" + columns
columns = ['household_key'] + list(columns)
dfstd.columns = columns

df = pd.merge(df, dfstd, how="left", on=['household_key'])

df
```

## Variação de sales_value por período

Como não temos o sales_value para o último período vou criar uma variável de taxa de variação defasada em 1 tempo

```python
df["SALES_VALUE_var1"] = df.groupby(["household_key"])['SALES_VALUE'].shift(1) / df.groupby(["household_key"])['SALES_VALUE'].shift(2)
df["SALES_VALUE_var2"] = df.groupby(["household_key"])['SALES_VALUE'].shift(2) / df.groupby(["household_key"])['SALES_VALUE'].shift(3)
df["SALES_VALUE_var3"] = df.groupby(["household_key"])['SALES_VALUE'].shift(3) / df.groupby(["household_key"])['SALES_VALUE'].shift(4)
df = df.replace(np.inf, np.nan)

df.head(20)
```

# Modelagem

## Xgboost

```python
df = df.drop(columns = ['household_key', 'WEEK_NO_BINNED','COUPON_DISC','COUPON_DISC_BOOL'])
```

```python
y = df.loc[:,'SALES_VALUE']
X = df.drop(columns=['SALES_VALUE'])
```

```python
df.columns
```

```python
df.head()
```

```python
# Separar a base em treino e teste para validaçao

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
import xgboost
from xgboost import XGBRegressor
```

```python
# Criando o modelo

model_xgb = XGBRegressor()
model_xgb.fit(X_train, y_train)
```

```python
# Prediçao

predic = model_xgb.predict(X_test)

predic
```

```python
# Criando um DataFrame com o SALES_VALUE e suas prediçoes

df2 = pd.DataFrame({'SALES_VALUE':y_test, 'Resultado':predic})
df2
```

### Validação por r^2

```python
# Plotando o Grafico do preço em relaçao as prediçoes
import matplotlib.lines as mlines
import matplotlib.transforms as mtransforms

fig, ax = plt.subplots()
sns.regplot(x="SALES_VALUE", y="Resultado", data=df2);
sns.scatterplot(x='SALES_VALUE',y='Resultado', data=df2)
line = mlines.Line2D([0, 1], [0, 1], color='red')
transform = ax.transAxes
line.set_transform(transform)
ax.add_line(line)
plt.show()
```

```python
from scipy import stats

slope, intercept, r_value, p_value, std_err = stats.linregress(y_test ,predic)

print ("r-squared:", r_value**2)
```

### Validação por RMSE (cross-val)

Aqui utilizei cross validation para validar o modelo, porém como a série é temporal e usei defasagens deveria ter levado em consideração o contexto temporal. A forma que utilizei está incorreta, porém meu objetivo era só comparar os modelos e não validar os valores reais do RMSE e r^2

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedKFold
from numpy import absolute

cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)

scores = cross_val_score(model_xgb, X, y, scoring='neg_root_mean_squared_error', cv=cv, n_jobs=-1)

scores = absolute(scores)
print('RMSE: %.3f (%.3f)' % (scores.mean(), scores.std()) )
```
